```python
def print_items(n)
  for i in range(n):
      print (i)
print_items (10)
```

```python
#Big O(2n) [n + n = 2n] {Time Complexity}
def print_items(n):
    for i in range(n):
        print (i)
    for j in range(n):
        print (j)
print_items(10)
```

```python
#Big O(n^2) {Time Complexity}
def print_items(n):
    for i in range(n):
        for j in range(n):
            print (i,j)
print_items(10)
```

```python
#Big [O(n^2) + O(n)]
def print_items(n):
    for i in range(n):
        for j in range(n):
            print (i,j)
    for k in range (n):
        print (k)
print_items(10)
```

```python
#Big O(1) [Constant Time]
def add_items(n):
    return n + n
add_items (20)
```
